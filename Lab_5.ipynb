```python
import pandas as pd
```

```python
toluca = pd.read_csv("toluca.csv")
credit = pd.read_csv("Credit.csv")
```

```python
import matplotlib as plt
%matplotlib inline

toluca.plot(kind='scatter', x='lotSize', y='workHours')
```

```python
import seaborn as sns

r1 = sns.residplot('lotSize', 'workHours', data=toluca)
r1.set(xlabel='Lot Size', ylabel='Residual')
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(toluca.drop('workHours', axis=1), toluca.iloc[:,1])
```

```python
reg.coef_
```

```python
reg.intercept_
```

Therefore the line would be y = 3.57x + 62.36. But Lotsize doesn't seem a good predictor with the line.

# Plots for credit dataset 

## Limit vs Income

```python
import matplotlib as plt
%matplotlib inline

credit.plot(kind="scatter", y="Limit", x="Income")
```

## Limit vs Residual

```python
import seaborn as sns

r1 = sns.residplot('Limit', 'Income', data=credit)
r1.set(xlabel='Limit', ylabel='Residual')
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(credit[['Income']], credit[['Limit']])
reg.coef_, reg.intercept_
```

Therefore we see that the Income can be a predictor for Limit with line y = 51.875x + 2389.869
